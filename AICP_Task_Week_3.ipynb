Q.1: Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis
and a title.

Q.2: Write a Python program to plot two or more lines with legends, different widths and colors.
The code snippet gives the output shown in the following screenshot:


```python
x_1 = [10, 20, 40]
y_1 = [20, 40, 10]
x_2 = [10, 20, 30]
y_2 = [35, 15, 25]

plt.plot(x_1, y_1, color='blue', linewidth=3, label='line1-width-3')
plt.plot(x_2, y_2, color='red', linewidth=5, label='line2-width-5')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two Lines with Different Widths')

plt.legend()
plt.show()
```

Q.4: Write a Python program to display a horizontal bar chart of the popularity of programming
Languages. Here is sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:


```python
lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
pop = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(lang, pop, color='darkgreen')

plt.xlabel('Popularity')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages')

plt.grid(True)
plt.show()
```

Q.5: Write a Python program to create bar plot from a DataFrame. Sample Data Frame:
a b c d e
2 4,8,5,7,6
4 2,3,4,2,6
6 4,7,4,7,8
8 2,6,4,8,6
10 2,4,3,3,2
The code snippet gives the output shown in the following screenshot:

```python
data = {'a': [2, 4, 6, 8, 10],
        'b': [4, 2, 4, 2, 2],
        'c': [8, 3, 7, 6, 4],
        'd': [5, 4, 4, 8, 3],
        'e': [7, 2, 7, 6, 2]}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'cyan', 'purple']

df.plot(kind='bar', color=colors)
plt.grid(True)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q.6: Write a Python program to create a pie chart of gold medal achievements of five most successful
countries in 2016 Summer Olympics. Read the data from a csv file.
Sample data:
country,gold_medal
United States,46
Great Britain,27
China,26
Russia,19
Germany,17

```python
df = pd.read_csv("olympics_medals.csv")
sorted_data = df.sort_values(by='summer_gold', ascending=False).head(5)

ct = sorted_data['countries '].tolist()
gm = sorted_data['summer_gold'].tolist()

explode = (0.3, 0, 0, 0, 0)
plt.figure(figsize=(8, 8))

plt.pie(gm, labels=ct, autopct='%1.1f%%', startangle=150, explode=explode, shadow=True)
plt.title('Gold Medal Achievements (Top 5 Countries)')
plt.axis('equal')
plt.show()
```
